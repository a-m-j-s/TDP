```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sdas.tests.LoadSdasData import LoadSdasData
from sdas.tests.StartSdas import StartSdas
```

```python
# Load and select data
shot = 38566
client = StartSdas()
V_data, time = LoadSdasData(
    client, 'PCIE_ATCA_ADC_16.BOARD_1.CHANNEL_026', shot)
I_data, _ = LoadSdasData(
    client, 'PCIE_ATCA_ADC_16.BOARD_1.CHANNEL_028', shot)
IP_data,time2 = LoadSdasData(
    client,'MARTE_NODE_IVO3.DataCollection.Channel_100',shot)

bool_indices = (time > 125750) & (time < 129750)
bool_indices2 = (time2 > 125750) & (time2 < 129750)

V_data = V_data[bool_indices]
I_data = I_data[bool_indices]
IP_data = IP_data[bool_indices2]
time = time[bool_indices]
time2 = time2[bool_indices2]
```

```python
#Draw data
fig = plt.figure(frameon=True)
ax = fig.add_axes([0,0,1,1])
ax.plot(time2/1000 ,IP_data )
ax.set_ylim(0,4500)
ax.set_xlabel('time (ms)')
ax.set_ylabel('Plasma Current (mA)')
ax.set_title('Plasma Current plot')
fig.savefig('PlasmaCurrent94mm.png',bbox_inches='tight',pad_inches=0.1)
```

```python
def func(x, a, b, c):
    return a * (1-np.exp((x-b)/c))
```

```python
#function that does fit

def fit(dataV,dataI,vmax,f):
    bool_indices = (dataV<vmax)
    dataV_fit=dataV[bool_indices]
    dataI_fit=dataI[bool_indices]

    parameters, covariance = curve_fit(f, dataV_fit, dataI_fit)
    return parameters,covariance
```

```python
#loop over various voltages

Vmaxarray=[5,7.5,10,12.5,15,17.5,20,22.5,25,27.5,30,32.5,35]
parametersarray=[]
covariancearray=[]
for v in Vmaxarray:
    p,c=fit(V_data,I_data,v,func)
    parametersarray.append(p)
    covariancearray.append(c)
```

```python
Tearray=[parametersarray[i][2] for i in range(0,len(Vmaxarray))]
Teerrorarray=[covariancearray[i][2][2] for i in range(0,len(Vmaxarray))]
```

```python
Vfarray=[parametersarray[i][1] for i in range(0,len(Vmaxarray))]
Vferrorarray=[covariancearray[i][1][1] for i in range(0,len(Vmaxarray))]
```

```python
Isatarray=[parametersarray[i][0] for i in range(0,len(Vmaxarray))]
Isaterrorarray=[covariancearray[i][0][0] for i in range(0,len(Vmaxarray))]
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.errorbar(Vmaxarray, Tearray,yerr=Teerrorarray,xerr=None,ecolor='Black',capsize=4)
ax.set_xlabel('Cutoff Voltage (V)')
ax.set_ylabel('Electron Temperature $(eV)$')
fig.savefig('Cutoff94mm.png',bbox_inches='tight',pad_inches=0.1)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.errorbar(Vmaxarray, Vfarray,yerr=Vferrorarray,xerr=None,ecolor='Black',capsize=4)
ax.set_xlabel('Cutoff Voltage (V)')
ax.set_ylabel('Floating Potential $(V)$')
fig.savefig('CutoffVf94mm.png',bbox_inches='tight',pad_inches=0.1)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.errorbar(Vmaxarray, Isatarray,yerr=Isaterrorarray,xerr=None,ecolor='Black',capsize=4)
ax.set_xlabel('Cutoff Voltage (V)')
ax.set_ylabel('Saturation Current $(mA)$')
fig.savefig('CutoffIsat94mm.png',bbox_inches='tight',pad_inches=0.1)
```

```python
e=1.602e-19 #C
k=1.381e-23 #J K^-1
m_i=1.6726219e-27 #kg proton mass
A=6e-6 #m^2
Tearray=[parametersarray[i][2]*e/(k) for i in range(2,7)]
Isatarray=[parametersarray[i][0]*0.001 for i in range(2,7)]
Vfarray=[parametersarray[i][1] for i in range(2,7)]
csarray=[np.sqrt(2*k*Tearray[i]/m_i) for i in range(0,5)]
nearray=[2*Isatarray[i]/(e*A*csarray[i]) for i in range(0,5)]
Vparray=[Vfarray[i]+3*k*Tearray[i]/e for i in range(0,5)]

print("Te = {:.5E}eV".format(np.mean(Tearray)*8.61733e-5))
print("Vf ={:.5E} V".format(np.mean(Vfarray)))
print("Vp ={:.5E} V".format(np.mean(Vparray)))
print("cs ={:.5E} m/s".format(np.mean(csarray)))
print("n ={:.5E} m^-3".format(np.mean(nearray)))
print("Tesigma = {:.2E}eV".format(np.std(Tearray)*8.61733e-5))
print("Vfsigma ={:.2E} V".format(np.std(Vfarray)))
print("Vpsigma ={:.2E} V".format(np.std(Vparray)))
print("nsigma ={:.2E} m^-3".format(np.std(nearray)))
print("cssigma ={:.2E} m/s".format(np.std(csarray)))
print("Isat ={:.5E} V".format(np.mean(Isatarray)*1000))
print("Isatsigma ={:.2E} m/s".format(np.std(Isatarray)*1000))
```

```python
# calculation of the other parameters

e=1.602e-19 #C
k=1.381e-23 #J K^-1
m_i=1.6726219e-27 #kg proton mass
I_sat=parametersarray[3][0]*0.001
I_sat_var=covariancearray[3][0][0]*0.001
V_f=parametersarray[3][1]
V_f_var=covariancearray[3][1][1]
T_e=parametersarray[3][2]*e/k
T_e_var=covariancearray[3][2][2]*e/k
c_s=np.sqrt(2*k*T_e/m_i)
c_s_var=k/m_i*1/np.sqrt(2*k*T_e/m_i)*T_e_var
A=6e-6 #m^2
n_e=2*I_sat/(e*A*c_s)
n_e_var=np.sqrt((2*I_sat_var/(e*A*c_s))*(2*I_sat_var/(e*A*c_s))+(2*I_sat*c_s_var/(e*A*c_s*c_s))*(2*I_sat*c_s_var/(e*A*c_s*c_s)))
V_p=V_f+3*k*T_e/e
V_p_var=np.sqrt(V_f_var*V_f_var+(3*k*T_e_var/e)*(3*k*T_e_var/e))

print("Te = {:.5E}eV".format(T_e*8.61733e-5))
print("Vf ={:.5E} V".format(V_f))
print("Vp ={:.5E} V".format(V_p))
print("cs ={:.5E} m/s".format(c_s))
print("n ={:.5E} m^-3".format(n_e))
print("Tesigma = {:.2E}eV".format(T_e_var*8.61733e-5))
print("Vfsigma ={:.2E} V".format(V_f_var))
print("Vpsigma ={:.2E} V".format(V_p_var))
print("nsigma ={:.2E} m^-3".format(n_e_var))
print("cssigma ={:.2E} m/s".format(c_s_var))
```
